#Part 1 - Create histogram of 5000 datasets mean values

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def set_seed(seed):
    #print("Seed value:", seed)
    np.random.seed(seed)
```

```python
def create_dataset(lam, size):
    dataset = np.random.poisson(lam=lam, size=size)
    #print("Generated dataset:", dataset)
    return dataset
```

```python
def find_mean(dataset):
    mean = np.mean(dataset)
    #print("Mean value:", mean)
    return mean
```

```python
random_seed = 10
random_poison_mean = 3
random_sample_size = 30
```

```python
def create_mean_values_histogram(mean=3, sample_size=30, bins=None):
  my_list_of_means = []
  templist = []
  #print ("My mean value", mean)
  #print ("My sample size", sample_size)
  #print ("my bin value", bins)

  if random_seed is not None:
    set_seed(random_seed)

  for _ in range(5000):
    temp_dataset = create_dataset(mean,sample_size)
    temp_mean = find_mean(temp_dataset)
    my_list_of_means.append(temp_mean)

  #print("List of mean values", my_list_of_means)

  # Plot histogram
  if bins is None:
    plt.hist(my_list_of_means, density=True, alpha=0.7, color='blue', edgecolor='black')
  else:
    plt.hist(my_list_of_means, bins=bins, density=True, alpha=0.7, color='blue', edgecolor='black')

  #plt.hist(my_list_of_means, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')
  plt.title('Histogram of 5000 samples mean')
  plt.xlabel('Mean Value')
  plt.ylabel('Frequency')
  plt.show()
```

```python
create_mean_values_histogram(mean=3, sample_size=50)
create_mean_values_histogram(mean=3, sample_size=40, bins=30)
create_mean_values_histogram(mean=3, sample_size=30, bins=60)
```

# Part 2 - Dictionary Practice

```python
import urllib.request
import string
```

```python
# a. Read raw data from web
url = "https://www4.stat.ncsu.edu/~online/datasets/charles-dickens-oliver-twist.txt"
urllib.request.urlretrieve(url, 'charles-dickens-oliver-twist.txt')
```

```python
# now that we have a reference to the url, read the data into a string
with open('charles-dickens-oliver-twist.txt', 'r') as f:
    lines = f.read()
```

```python
# b. Print the first 1000 characters and last 1000 characters
print("First 1000 characters ----> \n", lines[:1000])
print("Last 1000 characters  ----> \n", lines[-1000:])
```

```python
# c. Subset the string to remove front and end matter
find_start_index = lines.find("CHAPTER I")
find_end_index = lines.find("End of the Project Gutenberg EBook")
my_book = lines[find_start_index:find_end_index].strip()

print("First 1000 characters ----> \n", my_book[:250])
print("\n\n\n")
print("Last 1000 characters  ----> \n", my_book[-250:])
```

```python
# d. Count the occurrences of each word using a dictionary
word_dictionary = {}
temp = 0
# For testing purposes. Comment the next line when running the program.
# my_book = "This is just for test-ing purposes. \n Saurabh is leaning dictionarie's & string function's in python programming!"

# Replace '\n' with space
my_book_without_newline = my_book.replace('\n', ' ')

# Split the string by spaces
word_list = my_book_without_newline.split()

for word in word_list:
    word = word.lower()
    for punctuation_char in string.punctuation:
        word = word.replace(punctuation_char, '')
    if word != '':
        if word in word_dictionary:
            word_dictionary[word] = word_dictionary[word] + 1
        else:
            word_dictionary[word] = 1
    else:
        temp = temp + 1
```

## stack overflow reference
https://stackoverflow.com/questions/20577840/python-dictionary-sorting-in-descending-order-based-on-values


A lambda function is an anonymous function. In this case, lambda kv: kv[1] takes a parameter kv (which represents each item in the iterable), and it returns kv[1]. This means that the lambda function extracts the second element of each tuple in the form of (key, value).

word_dictionary.items(): This returns a view of the dictionary's key-value pairs as tuples, like [(key1, value1), (key2, value2), ...].
key=lambda kv: kv[1]: This specifies that the sorting should be based on the second element of each tuple (the values).
reverse=True: This indicates that the sorting should be in descending order.


```python
# e. print out the 40 most prevalent words and how prevalent they were
#sorted_word_dictionary = dict(sorted(word_dictionary.items(), reverse=True))
#sorted_word_dictionary = sorted(word_dictionary, key=word_dictionary.get, reverse=True)
my_dictionay = sorted(word_dictionary.items(), key=lambda kv: kv[1], reverse=True)

print("Top 40 words and the counts  ---->  \n")

for i in range(40):
  print(my_dictionay[i])
```

```python
# f. Remove the following words from dictionary and then print the 40 most common words:
# my_dictionay is a list. It is no more a dictionary. So remove the common word from orginal dictonary and than again sort by value

words_to_remove = ['the', 'and', 'to', 'of', 'a', 'in', 'that', 'with', 'for', 'at', 'on']
for word in words_to_remove:
    word_dictionary.pop(word, None)

my_dictionay = sorted(word_dictionary.items(), key=lambda kv: kv[1], reverse=True)
print("Top 40 words after removing common words  ---->  \n")

for i in range(40):
  print(my_dictionay[i])
```
